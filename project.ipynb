```python
directory = "C:\\Users\\lenovo\\Desktop\\AI Design Project\\Code\\dataset"
vehicle_dataset = tf.keras.utils.image_dataset_from_directory(
                    directory,
                    labels="inferred",
                    label_mode="int",
                    class_names=class_names,
                    color_mode="rgb",
                    batch_size=batch_size,
                    image_size=(image_weight,image_width),
                    shuffle=True,
                    seed=None,
                    validation_split=None,
                    subset=None,
                    interpolation="bilinear",
                    follow_links=False,
                    crop_to_aspect_ratio=False
)
```
